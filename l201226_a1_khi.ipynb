```python
id_ = []##
name = []##
designation = []##
hec_approved = []##
highest_qualification = [] 
email = []##
department = []
extension = []  ##
profile = []##
img = []##
```

```python
for i in range(len(links)):
    print ("hello",i)
    if i == 0:
        dept = 'cs'
    elif i == 1:
        dept = 'cyber'
    elif i == 2:
        dept = 'ai'
    elif i == 3:
        dept = 'se'
    elif i == 4:
        dept = 'ee'
    elif i == 5:
        dept = 'fsm'
    elif i == 6:
        dept = 'ss'
    else:
        dept = 'unknown'
    url_faculty=links[i] #0=cs faculty 
    r=requests.get(url_faculty)
    soup_faculty=BeautifulSoup(r.content, 'html.parser')
    x=soup_faculty.find_all('div', class_='gdlr-core-personnel-list-column')
    length=len(x)
    for faculty in x:
        print(dept)
        department.append(dept)
        try:
            name.append(faculty.find('h3').text)#name
        except:
            name.append('Not Found')
        try:
            profile.append((faculty.find('a')['href']))#link
        except:
            profile.append('Not Found')
        try:
            designation.append(faculty.find('div', class_='gdlr-core-personnel-list-position').text)#position
        except:
            designation.append('Not Found')
        try:
            email.append(faculty.find('div', class_='kingster-personnel-info-list').text)#email
        except:
            email.append('Not Found')
        try:
            extension.append(faculty.find("div", class_="kingster-personnel-info-list").find_next_sibling("div").text)#phone
        except:
            extension.append('Not Found')
        #give link to image
        try:
            image = faculty.find('img')
            if image is not None:
                image_src = image['src']
                img.append(image_src)
            else:
                img.append("not found")
        except:
            img.append("not found")
        length=length-1
        print(length)
        print('------------------')
    #==================================================================================================
    
```

```python
for i in range(len(profile)):
        url=profile[i]
        url
        r=requests.get(url)
        soup=BeautifulSoup(r.content, 'html.parser')
        print("in proifle: ",i)
        try:
                x=soup.find('span',class_='gdlr-core-title-item-caption').text.split('\n')[2]#hec approved
                if x:
                    hec_approved.append('True')
                else:
                    hec_approved.append('False')
            # print(soup.find('span',class_='gdlr-core-title-item-caption').text.split('\n')[2])#hec approved
            # print("end of hec ")
        except:
            print('Not Found')
            print("end of hec ")
            hec_approved.append('False')
        try:
            highest_qualification.append(soup.find('span',class_='gdlr-core-icon-list-content').get_text())
        except:
            highest_qualification.append('Not Found')
            print("end of edu ")
        try:
            script_block = soup.find('script', id='burst-js-extra')
            if script_block:
                script_content = script_block.string
                page_id_match = re.search(r'"page_id":"(\d+)"', script_content)
                if page_id_match:
                    page_id = page_id_match.group(1)
                    print("Extracted page_id:", page_id)
                    print("end of id ")
                    id_.append(int(page_id))
                else:
                    print("Page ID not found in the script content.") 
                    id_.append(-1)     
        except:
            id_.append('Not Found')     
            continue
        print('--------',i,'----------')
```
