```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
file_path = 'FAST NUCES Islamabad.html'
# response = requests.get(URL) 
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()
soup = BeautifulSoup(html_content, 'html.parser')
print(soup.prettify())
```

```python
#<div class="col-lg-3 col-md-6 col-sm-6" style="margin-top:20px;"><div class="our-team"><div class="pic"><img id="myImage4553" src="./FAST NUCES Islamabad_files/4553.jpg"></div><div class="team-content"><h3 class="title">Waseem Shahzad</h3><span class="post">Director &amp; Professor</span> <p>waseem.shahzad@nu.edu.pk</p><br></div><ul class="social"><li><a href="http://isb.nu.edu.pk/Faculty/Details/4553" target="_blank" style="font-style: italic;">More...</a></li></ul></div></div>
faculty=soup.find_all("div",class_='our-team')
print(len(faculty))
```

```python
id_ = []##
name = []##
designation = []##
hec_approved = []##
highest_qualification = [] 
email = []##
department = []##
extension = []  ##
profile = []##
img = []##
```

```python
for person in faculty:
   try:
      img_info=person.find("img")
      image=img_info['src']
      img.append(image)
   except:
      img.append("Not Found")
   
   try:
      name.append(person.find("h3",class_='title').text)
   except:
      name.append("Not Found")
      
   try:
      designation.append(person.find("span",class_='post').text)
   except:
      designation.append("Not Found")
      
   try:
      email.append(person.find("p").text)
   except:
      email.append("Not Found")
      
   try:
      profile.append(person.find("ul",class_='social').find("li").find("a")['href'])
   except:
      profile.append("Not Found")
   try:
      image=person.find("img")['id']
      id_.append(int(image[-4:]))
   except:
      id_.append(-1)
   
         
   
   
print("len of img",len(img))
print("len of name",len(name))
print("len of designation",len(designation))
print("len of email",len(email))
print("len of profile",len(profile))
print("len of id",len(id_))
```

```python
for i in range(len(profile)):
    url=profile[i]
    url
    r=requests.get(url)
    soup=BeautifulSoup(r.content, 'html.parser')
    print("in proifle: ",i)   
    
    try:        
        # print(soup.find('ol',class_='breadcrumb').find_all('li',class_='breadcrumb-item')[1].text.strip())
        department.append(soup.find('ol',class_='breadcrumb').find_all('li',class_='breadcrumb-item')[1].text.strip())
    except:
        department.append("Not Found")
    try:
        found=soup.find('label', text='HEC Approved PhD Supervisor')
        if found:
            print("hec")
            hec_approved.append("True")
        else:
            print("not hec")
            hec_approved.append("False")
    except:
            hec_approved.append("Not Found")
    
    try:
        highest_qualification.append('Not Found')
    except:
        highest_qualification.append("Not Found")
    try:
        extension.append((soup.find('i', class_='fa fa-phone').findNext('span').text.strip()[-3:]))
    except:
        extension.append("Not Found")
        
```

```python
print("Count of 'id_':", len(id_))
print("Count of 'name':", len(name))
print("Count of 'designation':", len(designation))
print("Count of 'hec_approved':", len(hec_approved))
print("Count of 'email':", len(email))
print("Count of 'department':", len(department))
print("Count of 'img':", len(img))
print("Count of 'profile':", len(profile))
print("Count of 'highest_qualification':", len(highest_qualification))
print("Count of 'extension':", len(extension))
```
