```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://lhr.nu.edu.pk/faculty/'
r=requests.get(url)
print(r)

```

```python
soup=BeautifulSoup(r.content, 'html.parser')
# print(soup)
print(soup.prettify())
```

```python
soup.findAll('div',class_='facultyCard')
```

```python
#how many cards are there:
length=len(soup.findAll('div',class_='facultyCard'))
```

```python
id_ = []
name = []
designation = []
hec_approved = []
highest_qualification = []  
email = []
department = []
extension = []  
profile = []
img = []
```

```python
for i in range(length):
    faculty = soup.findAll('div', class_='facultyCard')[i]
    
    try:
        id_.append(int(faculty.find(class_='faculty-link')['href'].split('/')[-1]))
    except:
        id_.append(-1)
    
    try: 
        name.append(faculty.find('h5', class_='text-center').text)
    except:
        name.append('Not Found')
    
    designation_info = faculty.find('p', class_='small')
    if designation_info:
        designation_text = designation_info.text.strip().split('\n')[0]
    else:
        designation_text = 'Not Found'
    designation.append(designation_text)
    
    try:
        hec_approved.append(bool((faculty.find('p',class_='small').text.strip().split('\n')[2]=='                          HEC Approved PhD Supervisor')))
    except:
        hec_approved.append(False)

    email_info = faculty.find('p', class_='mb-0')
    if email_info:
        email_text = email_info.text
    else:
        email_text = 'Not Found'
    email.append(email_text)
    
    department_element = faculty.find(class_='faculty-link')
    if department_element:
        department_text = department_element['href'].split('/')[1]
    else:
        department_text = 'Not Found'
    department.append(department_text)

    img_info = faculty.find('img', class_='card-img-top')
    if img_info:
        img_url = 'https://lhr.nu.edu.pk/' + img_info['src']
    else:
        img_url = 'Not Found'
    img.append(img_url)

    profile_url_info = faculty.find('a', class_="faculty-link")
    if profile_url_info:
        profile_url = "https://lhr.nu.edu.pk/" + profile_url_info["href"]
    else:
        profile_url = 'Not Found'
    profile.append(profile_url)

```

```python
#for all 
for url2  in profile:
    r2=requests.get(url2)
    soup2=BeautifulSoup(r2.content, 'html.parser')
    # print(soup2.prettify())
    print(extension,"b4 try catch")
    try:
        extension.append(int(soup2.find('span',class_='small').text.split(':')[-1]))
        # print(extension,"during try catch")
        print(" ",len(extension))
    except:
        extension.append('Not Found')
    try:
        highest_qualification.append(soup2.find('div',class_='col-lg-8 col-md-6 col-sm-12 text-justify').find('ul').find_all('li')[0].text)
        print(highest_qualification,"during try catch")
        print(" ",len(highest_qualification))
    except:
        highest_qualification.append('Not Found')
```

```python


print("Count of 'id_':", len(id_))
print("Count of 'name':", len(name))
print("Count of 'designation':", len(designation))
print("Count of 'hec_approved':", len(hec_approved))
print("Count of 'email':", len(email))
print("Count of 'department':", len(department))
print("Count of 'img':", len(img))
print("Count of 'profile':", len(profile))
print("Count of 'highest_qualification':", len(highest_qualification))
print("Count of 'extension':", len(extension))

```

```python
df = pd.DataFrame({
    'ID': id_,
    'Name': name,
    'Designation': designation,
    'HEC_Approved': hec_approved,
    'Highest_qualification': highest_qualification,  # You need to define how to extract this information
    'Email': email,
    'Department': department,
    'Extension': extension,  # You need to define how to extract this information
    'Profile': profile,
    'Image': img
})

df.to_csv('faculty.csv', index=False)
df
```
